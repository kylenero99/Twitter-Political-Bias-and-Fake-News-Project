```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ted_cruz_gw = pd.read_json("tedcruztweets-congress115-congress16.json", lines=True)
ted_cruz_gw
```

```python
ted_cruz_tweets_text_list = ted_cruz_gw["full_text"].tolist()
ted_cruz_hashtags = []

for tweet in ted_cruz_tweets_text_list:
    words = tweet.split(' ')
    for word in words:
        if len(word) > 0 and word[0] == "#":
            ted_cruz_hashtags.append(word)

ted_cruz_hashtags_as_df = pd.DataFrame(ted_cruz_hashtags, columns=["Hashtag"])
ted_cruz_hashtags_frequency = ted_cruz_hashtags_as_df.groupby("Hashtag").size().reset_index(name='counts').sort_values("counts", ascending=False)
ted_cruz_hashtags_frequency.head(20)
```

```python
plt.bar(ted_cruz_hashtags_frequency['Hashtag'][:4], ted_cruz_hashtags_frequency['counts'][:4])
```
