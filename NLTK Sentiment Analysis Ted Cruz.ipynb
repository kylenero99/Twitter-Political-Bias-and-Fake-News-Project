```python
import pandas as pd
```

```python
df = pd.read_json("tedcruztweets-congress115-congress16.json", lines = True)
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
test = sid.polarity_scores(df.full_text.tolist()[10])
print(df.full_text.tolist()[10])
print(test)
```

```python
scores = []
for i in df.full_text.tolist():
    scores.append(sid.polarity_scores(i))
df["polarity_scores"] = scores
df['compound']  = df['polarity_scores'].apply(lambda score_dict: score_dict['compound'])
res = []
for j in df.compound.tolist():
    if j > 0:
        res.append('positive')
    elif j == 0:
        res.append('neutral')
    elif j < 0:
        res.append("negative")
df["comp_score"] = res
df.head()
```

```python
positive_tweets = df.loc[df.comp_score == "positive"]
positive_tweets = positive_tweets.sort_values(by = 'compound', ascending = False)
print(str(round((len(positive_tweets) / len(df)) * 100, 2)) + "%")
positive_tweets.head()
```

```python
negative_tweets = df.loc[df.comp_score == "negative"]
negative_tweets = negative_tweets.sort_values(by = 'compound', ascending = True)
print(str(round((len(negative_tweets) / len(df)) * 100, 2)) + "%")
negative_tweets.head()
```

```python
neutral_tweets = df.loc[df.comp_score == "neutral"]
print(str(round((len(neutral_tweets) / len(df)) * 100, 2)) + "%")
neutral_tweets.head()
```
